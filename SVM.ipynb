```python
from sklearn.datasets import load_iris
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
```

```python
iris = load_iris(as_frame=True)
iris['frame'].describe()
```

```python
x = iris.data[['petal length (cm)','petal width (cm)']].values   #iris.data特徵
y = (iris.target ==2)         #iris.target標籤

svm_clf = make_pipeline(StandardScaler(),LinearSVC(C=1,random_state=42))  #找到一個最佳的線性分隔超平面
print(svm_clf)
svm_clf.fit(x,y)

x_new = [[5.5,1.7],[5.0,1.5]]
print(svm_clf.predict(x_new))

print(svm_clf.decision_function(x_new))
```

```python
#非線性SVM
from sklearn.datasets import make_moons
from sklearn.preprocessing import PolynomialFeatures

x,y = make_moons(n_samples=100,noise=0.15,random_state=42)
polynomial_svm_clf = make_pipeline(PolynomialFeatures(degree=3),StandardScaler(),LinearSVC(C=10,max_iter=10_000,random_state=42))
polynomial_svm_clf.fit(x,y)

x_new = [[5.5,1.7],[5.0,1.5]]
print(polynomial_svm_clf.predict(x_new))

print(polynomial_svm_clf.decision_function(x_new))
```

```python
#kernel技巧
from sklearn.svm import SVC

poly_kernel_svm_clf = make_pipeline(StandardScaler(),SVC(kernel='poly',degree=3,coef0=1,C=5))
poly_kernel_svm_clf.fit(x,y)

x_new = [[5.5,1.7],[5.0,1.5]]
print(poly_kernel_svm_clf.predict(x_new))

print(poly_kernel_svm_clf.decision_function(x_new))
```

```python
#相似度特徵
#Gaussian RBF
rbf_kernel_svm_clf = make_pipeline(StandardScaler(),SVC(kernel='rbf',gamma=5,C=0.001))
rbf_kernel_svm_clf.fit(x,y)

x_new = [[5.5,1.7],[5.0,1.5]]
print(poly_kernel_svm_clf.predict(x_new))

print(poly_kernel_svm_clf.decision_function(x_new))
```

```python
#SVM回歸
from sklearn.svm import LinearSVR
svm_reg = make_pipeline(StandardScaler(),LinearSVR(epsilon=0.5,random_state=42))
svm_reg.fit(x,y)
```
